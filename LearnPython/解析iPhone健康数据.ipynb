数据说明：
- type：表示数据的类型，可以能的值为
    - HKQuantityTypeIdentifierBodyMassIndex 体重指标
    - HKQuantityTypeIdentifierHeight 省高（cm）
    - HKQuantityTypeIdentifierBodyMass 体重（kg）
    - HKQuantityTypeIdentifierHeartRate 心率
    - HKQuantityTypeIdentifierStepCount 步数
    - HKCategoryTypeIdentifierSleepAnalysis 睡眠分析
    - HKQuantityTypeIdentifierFlightsClimbed 攀升高度
    - HKQuantityTypeIdentifierActiveEnergyBurned 燃脂
    - HKQuantityTypeIdentifierDistanceWalkingRunning 跑步
    
- sourceName：表示数据源，可能的值为
    - 小米运动
    - 健康
    - iPhone
 
- unit：表示记录值得单位，可能的是为
    - count
    - cm
    - kg
    - count/min
    - kcal
    - km
    
- sourceVersion：数据源版本   
- creationDate：创建日期
- startDate：记录开始时间
- endDate：记录结束时间
- value：值


````python
from xml.parsers.expat import ParserCreate

# 解析出想要的数据
def parse_data(name, attrs):
    print('sax:start_element: %s, attrs: %s' % (name, str(attrs)))
    # 解析 iphone 步数数据
    if name == 'HKQuantityTypeIdentifierStepCount' and attrs['sourceName'] == 'iPhone':
        value = attrs['value']
        creationDate = attrs['creationDate']
        startDate = attrs['startDate']
        endDate = attrs['endDate']
        step.write(value + ',' + creationDate + ','+ startDate + ','+ endDate)
    
    # 解析小米运动体重数据
    if name == 'HKQuantityTypeIdentifierBodyMass' and attrs['sourceName'] == '小米运动':
        value = attrs['value']
        creationDate = attrs['creationDate']
        startDate = attrs['startDate']
        endDate = attrs['endDate']
        bm.write(value + ',' + creationDate + ','+ startDate + ','+ endDate)

class DefaultSaxHandler(object):
    def start_element(self, name, attrs):
        #print('sax:start_element: %s, attrs: %s' % (name, str(attrs)))
        parse_data(name, attrs)

    def end_element(self, name):
        #print('sax:end_element: %s' % name)
        pass

    def char_data(self, text):
        #print('sax:char_data: %s' % text)
        pass


def parse_xml(xml):
    # 创建解析器
    handler = DefaultSaxHandler()
    parser = ParserCreate()
    # 处理函数
    parser.StartElementHandler = handler.start_element
    parser.EndElementHandler = handler.end_element
    parser.CharacterDataHandler = handler.char_data

    #解析
    parser.Parse(xml)


# 用于记录相关数据
step = open(r'E:\health\step.txt', 'a')
bm = open(r'E:\health\bm.txt', 'a')

# 读取 xml 文件
basefile = r'E:\health\export.xml'
with open(basefile, 'rb') as f:
    data = f.read()
    parse_xml(data)
    step.close()
    bm.close()
    
````
